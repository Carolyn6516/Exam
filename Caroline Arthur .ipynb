```python
#Question 3 
#Write a NUmPy program to compute the multiplication of two given matrixes. ([1,0],[0,1]) ([1,2,],[3,4])
import numpy as np
z=[[1,0],[0,1]]
y=[[1,2],[3,4]]
print("original matrix:")
print(z)
print(y)
result_1=np.dot(z,y)
print("Result of the said matrix multipication:")
print(result_1)
```

```python
#Question 4 
# Write a NUmPy program to create a random array with 1000 elements and compute the average, variance, standard deviation of the array elements.
import numpy as np
A = np.random.randn(1000)
print("Average of the array elements:")
mean = A.mean()
print(mean)
print("Standard deviation of the array elements:")
std = A.std()
print(std)
print("Variance of the array elements:")
var = A.var()
print(var)
```

```python
#Question 5 
#Write a Numpy program to calculate the cumulative sum of the elements along a given axis, sum over rows for each of the 3 columns and sum over columns for each of the 2 rows of a given 3x3 array.
import numpy as np
Z=np.array([[1,2,3], [4,5,6]])
print("Original array:")
print(Z)
print("Cumulative sum of the elements along a given axis:")
X=np.cumsum(Z)
print(X)
print("\nSum over rows for each of the 3 columns:")
X=np.cumsum(Z,axis=0)
print(X)
print("\nSum over columns for each of the 2 rows:")
X=np.cumsum(Z,axis=1)
print(X)
```

```python
import random

num = random.randint(1, 100)

attempts = 0
while True:
# Get input from the user
    guess = int(input("Guess the number (between 1 and 100): "))

# Increase the number of attempts
    attempts += 1

# Compare the user's guess to the system-generated number
    if guess == num:
        print("Congratulations! You guessed the number in", attempts, "attempts.")
        break
    elif guess <num:
        print("The number is higher. Try again.")
    else:
        print("The number is lower. Try again.")
```
